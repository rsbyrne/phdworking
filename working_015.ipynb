```python
import planetengine
import numpy as np
```

```python
from modelscripts import MS98X_systemscript as systemscript
from modelscripts import MS98X_observerscript as observerscript
```

```python
system = systemscript.build(res = 16)
```

```python
observer = observerscript.build()
```

```python
tempIC = planetengine.initials.sinusoidal.IC()
```

```python
cont_centre = 0.5
cont_width = 0.3
cont_thickness = 0.1
continent = (
    (cont_centre - cont_width / 2., 1. - cont_thickness),
    (cont_centre - cont_width / 2., 1.),
    (cont_centre + cont_width / 2., 1.),
    (cont_centre + cont_width / 2., 1. - cont_thickness),
    )
materialIC = planetengine.initials.extents.IC(vertexSets = ((1, continent),))
```

```python
initial = {'temperatureField': tempIC, 'materialVar': materialIC}
```

```python
model = planetengine.frame.Frame(
    system = system,
    observer = observer,
    initial = initial,
    instanceID = 'test'
    )
```

```python
model.report()
```

```python
model.checkpoint()
```

```python
model.iterate()
```

```python
model.checkpoint()
```

```python
model.load_checkpoint(0)
```
