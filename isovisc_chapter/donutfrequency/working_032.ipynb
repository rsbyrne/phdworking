```python
import planetengine
import os
from underworld import function as fn
import numpy as np
import math
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def chunks(l, n):
    """Yield successive n-sized chunks from l."""
    for i in range(0, len(l), n):
        yield l[i:i + n]
```

```python
def domFreqFunc(system):
    gradFunc = fn.math.dot(system.mesh.unitvec_r_Fn, system.temperatureField.fn_gradient)
    Nus = gradFunc.evaluate(system.outer)
    Nus = (Nus - np.average(Nus)) / np.average(Nus)
    sp = np.abs(np.fft.rfft(Nus, axis = 0))
    freq = np.fft.rfftfreq(Nus.size, 1. / (2. * sp.size))
    domFreq = int(max(zip(sp[:,0], freq))[1])
    return domFreq
```

```python
def NusProfile(system):
    gradFunc = fn.math.dot(system.mesh.unitvec_r_Fn, system.temperatureField.fn_gradient)
    Nus = gradFunc.evaluate(system.outer)
    Nus = (Nus - np.average(Nus)) / np.average(Nus)
    plt.plot(Nus)
    plt.show()
```

```python
modelfilenames = [filename for filename in os.listdir(outputPath) if filename[:6] == 'pemod_']
modelNames = [os.path.splitext(os.path.splitext(item)[0])[0] for item in modelfilenames]
```

```python
models = []
for modelName in modelNames:
    models.append(planetengine.frame.load_frame(outputPath, modelName, 'max'))
tosort = []
for model in models:
    tosort.append((model, model.params['Ra'], model.params['f']))
tosort = sorted(tosort, key=lambda tup: (tup[1], tup[2]))
models = [row[0] for row in tosort]
suiteDict = {}
for model in models:
    suiteDict[model.instanceID] = model
```

```python
models16 = []
models32 = []
models64 = []
for model in models:
    if model.params['res'] == 16:
        models16.append(model)
    elif model.params['res'] == 32:
        models32.append(model)
    elif model.params['res'] == 64:
        models32.append(model)
```

```python
rows = 7
cols = 8
models32grid = list(chunks(models32, rows))
```

```python
fig, axs = plt.subplots(rows, cols, figsize = (40, 40))
modelsgrid = list(chunks(models32, 7))
for row in range(rows):
    for col in range(cols):
        system = models32grid[col][row].system
        gradFunc = fn.math.dot(system.mesh.unitvec_r_Fn, system.temperatureField.fn_gradient)
        Nus = gradFunc.evaluate(system.outer)
        Nus = (Nus - np.average(Nus)) / np.average(Nus)
        axs[row][col].plot(Nus)
```

```python
for model in models32:
    NusProfile(model.system)
```

```python
domFreqData = {}
for model in models32:
    Ra = model.params['Ra']
    RaPow = int(math.log(Ra / 1e4, 2))
    if not RaPow in domFreqData:
        domFreqData[RaPow] = []
    f = model.params['f']
    domFreq = domFreqFunc(model.system)
    domFreqData[RaPow].append((f, domFreq))
for key in domFreqData:
    domFreqData[key] = sorted(domFreqData[key])
    domFreqData[key] = np.array(domFreqData[key]).transpose()
```

```python
def make_figs(models, figsDir):
    figs = []
    for model in models:
        name = model.instanceID
        fig = planetengine.visualisation.QuickFig(
            model.system.temperatureField,
            colourBar = False,
            edgecolour = 'black',
            facecolour = 'black',
            figsize = (400, 400),
            )
        figs.append(fig)
    figsDir = os.path.join('/workspace/user_data', figsDir)
    basefilename = 'fig_'
    for index, fig in enumerate(figs):
        num = str(index).zfill(3)
        filename = basefilename + num
        fig.save(os.path.join(figsDir, filename))
```

```python
make_figs(models16, 'data/donutfrequency/models16')
make_figs(models32, 'data/donutfrequency/models32')
```
