```python
import planetengine
import numpy as np
from planetengine.utilities import quickShow
```

```python
from modelscripts import MS98X_systemscript as systemscript
from modelscripts import MS98X_observerscript as observerscript
```

```python
system = systemscript.build(res = 64)
```

```python
observer = observerscript.build()
```

```python
tempIC = planetengine.initials.sinusoidal.IC()
```

```python
from planetengine import shapes
```

```python
continent = shapes.trapezoid(taper = 0.5, thickness = 0.02)
lowercontinent = shapes.trapezoid(taper = 0.5, thickness = 0.04)
verylowercontinent = shapes.trapezoid(taper = 0.5, thickness = 0.06)
basal = shapes.trapezoid(location = 'base', ratio = 0.5)
materialIC = planetengine.initials.extents.IC(
    vertexSets = (
        (3, verylowercontinent),
        (2, lowercontinent),
        (1, continent),
        (5, basal),
        )
    )
```

```python
initial = {'temperatureField': tempIC, 'materialVar': materialIC}
```

```python
planetengine.initials.apply(initial, system)
```

```python
quickShow(system.temperatureField, system.materialVar)
```
