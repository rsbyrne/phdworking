```python
import planetengine
from planetengine.utilities import quickShow

from modelscripts import MS98X_systemscript as systemscript
from modelscripts import MS98ann_handlerscript as handlerscript
from modelscripts import MS98X_initialscript as initialscript
```

```python
model = planetengine.model.Model(
    systemscript.build(
        res = 64,
        Ra = 1e7,
        surfT = 0.12,
        cont_buoyancy_mR = 10.,
        cont_creep_mR = 1.1,
        cont_creep_sR_mR = 1.1,
        cont_tau_mR = 0.9,
        cont_tau_bR_mR = 0.9,
        cont_heating_mR = 0.,
        cont_diffusivity_mR = 0.9,
        ),
    handlerscript.build(),
    initialscript.build(cont_thickness=0.2),
    outputPath = '../data',
    instanceID = 'kitchensink_002',
    )
```

```python
model.report()
```

```python
model.traverse(
    stopCondition = lambda: model.step.value > 10000,
    collectCondition = lambda: False, #model.step.value % 10 == 0,
    checkpointCondition = lambda: False, #model.step.value % 100 == 0,
    reportCondition = lambda: model.step.value % 100 == 0,
    )
```
