```python
import math
import random
```

```python
with open('../planetengine/moby.txt', 'r') as file:
    mobystring = file.read()
```

```python
print(mobystring[:100])
```

```python
splitherestring = mobystring.replace(' ', 'xxxsplitherexxx')
```

```python
str.isalpha
```

```python
cleanstring = ''.join(e for e in splitherestring if e.isalpha()).lower()
```

```python
rawmobylist = cleanstring.split('xxxsplitherexxx')
```

```python
mobylist = sorted(list(set(rawmobylist)))[1:]
```

```python
mobylist
```

```python
len(mobylist)
```

```python
def mobyhash(hashID):
    random.seed(hashID)
    mobyhashlen = math.ceil(len(hashID) * math.log(260, len(mobylist)))
    mobyhashlist = random.sample(mobylist, mobyhashlen)
    mobyhashstr = '_'.join(mobyhashlist).replace('__', '_').lower()
    return mobyhashstr
```

```python
hashID = '0dd5d59ee25ed58f1b27fd29858ffde7'
```

```python
mobyhash(hashID)
```
