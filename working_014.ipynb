```python
from modelscripts import MS98X_systemscript as systemscript
from planetengine.utilities import quickShow
from planetengine import initials
import numpy as np
```

```python
import planetengine
```

```python
system = systemscript.build()
```

```python
tempIC = initials.sinusoidal.Sinusoidal()
```

```python
cont_centre = 0.5
cont_width = 0.3
cont_thickness = 0.1
continent = np.array([
    [cont_centre - cont_width / 2., 1. - cont_thickness],
    [cont_centre - cont_width / 2., 1.],
    [cont_centre + cont_width / 2., 1.],
    [cont_centre + cont_width / 2., 1. - cont_thickness]
    ])
materialIC = initials.extents.Extents(volumes = (1, continent))
```

```python
tempIC.apply(system.temperatureField)
```

```python
materialIC.apply(system.materialVar)
```

```python
quickShow(system.temperatureField, system.materialVar)
```
