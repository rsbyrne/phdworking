```python
from planetengine import model
from planetengine import analysis
```

```python
modelnames = ['testsuite' + str(number).zfill(3) for number in range(10)]
```

```python
models = {}
for modelname in modelnames:
    models[modelname] = model.load_model(modelname)
    models[modelname].load_checkpoint(100000)
```

```python
testmodel = models['testsuite009']
```

```python
testmodel.data.analysers[0].analyse()
testmodel.data.analysers[0].dataDict['Nu']
```

```python
Nu = analysis.Analyse.Nu(testmodel.system.temperatureField, testmodel.system.mesh)
Nu.evaluate()
```
